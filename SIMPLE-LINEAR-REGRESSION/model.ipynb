```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

```python
# feature and target
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
# data points
plt.scatter(x,y, color = 'deepskyblue', s = 70, lw = 0.2)
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.show()
```

```python
# data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
# model train
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# pred
y_pred = lr.predict(x_test)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
# best fit line
plt.scatter(x,y, color = 'deepskyblue', s = 70, lw = 0.2, label = 'true data')
plt.plot(x_test, y_pred, color = 'k', label = 'best fit line')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.legend()
plt.show()
```

```python
# individual predictions
# user input

cgpa = float(input('Enter CGPA: '))
lr.predict([[cgpa]])
```
